- 탐색적 데이터 분석


# 데이터 종류
- 수치형 데이터
    - 연속형: 일정 범위 안에서 어떤 값이든 취할수 있는 실수형 데이터(온도, 키, 풍속 등)
    - 이산형: 횟수와 같은 정수형 데이터(사건의 발생빈도, 방개수 등)

- 범주형 데이터
    - 가능한 범주의 값만 취할수 있는 데이터
    - 성별: 남/여
    - 전공: 경영학/인문학/교육학(명목형)
    - 영화평점: 1/2/3/4/5(순서형)
    - 직급: 대리/과장/차장/부장(순서형)

- Feature, 독립변수, 설명변수
    - 학습 데이터의 특성
- class, label, target, 종속변수
    - 정답데이터

- 왜도(Skewness)
    - 데이터 분포의 비대칭도를 나타내는 통계량
    - 분포가 왼쪽으로 치우쳐져 있고, 오른쪽으로 긴 꼬리를 가지는 경우 왜도는 양수
    - 분포가 오른쪽으로 치우쳐져 있고, 왼쪽으로 긴 꼬리를 가지는 경우 왜도는 음수
    - 정규분포와 같이 좌우 대칭인 경우 왜도는 0에 가까워진다.

```python
df["fare"].skew()
```

```python
df["age"].skew()
```

```python
df.skew(numeric_only=True)
```

```python
df.count() # 결측치를 제외하고 count
```

```python
df.describe()
```

- 상관계수
    - 칼 피어슨(Karl Pearson)이 개발한 상관계수
    - 두개의 수치형 변수의 변화가 연관되는 정도
    - +1 ~ -1 사이의 값을 가짐
    - +1에 가까울수록 양의 상관관계
    - -1에 가까울수록 음의 상관관계
    - 0에 가까울수록 상관관계가 없음
    - 상관관계 != 인과관계

```python
df.iloc[:,1:].corr(numeric_only=True)
```
