### 평균과 분산 계산 방법

- **평균 ($\mu^u$)**: 발화 $u$의 전체 프레임 $N^u$에 대해 음성 특징 $x^u(n)$의 평균을 계산할 수 있습니다. 각 프레임의 특징 값들의 평균은 다음과 같이 정의됩니다.

$$
\mu^u = \frac{1}{N^u} \sum_{n=0}^{N^u - 1} x^u(n)
$$

- **분산 ($\sigma^{u^2}$)**: 분산은 각 프레임의 특징값 $x^u(n)$이 평균 $\mu^u$와 얼마나 차이가 나는지를 나타냅니다. 분산은 다음과 같이 정의됩니다.

$$
\sigma^{u^2} = \frac{1}{N^u} \sum_{n=0}^{N^u - 1} (x^u(n) - \mu^u)^2
$$

- **전체 발화에 대한 평균 ($\mu$)**: 발화 $u$에 대한 평균을 구한 후, 전체 발화에 대해 평균을 계산하려면 다음과 같이 서메이션을 사용하여 계산할 수 있습니다.

$$
\mu = \frac{1}{U} \sum_{u=0}^{U-1} \mu^u
$$

- **전체 발화에 대한 분산 ($\sigma^2$)**: 발화 $u$에 대한 분산을 구한 후, 전체 발화에 대해 분산을 계산하려면 다음과 같이 서메이션을 사용하여 계산할 수 있습니다.

$$
\sigma^2 = \frac{1}{U} \sum_{u=0}^{U-1} \sigma^{u^2}
$$

### 이식의 의미

- **평균**은 특정 발화의 특징값이 전반적으로 어떤 경향을 보이는지 나타내며, 특정 발화의 음성 특징이 전체 프레임을 통틀어 얼마나 높은지 또는 낮은지를 알 수 있습니다.
- **분산**은 발화 내에서 각 프레임의 특징값들이 평균값을 기준으로 얼마나 퍼져 있는지를 나타냅니다.
- **전체 발화에 대한 평균과 분산**은 여러 발화에서 공통적인 경향을 파악할 수 있도록 도와줍니다.
