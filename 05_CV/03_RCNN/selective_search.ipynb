```python
import cv2
import selective_search
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# 이미지 로드
image_path = "animals.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Selective Search (Fast Mode) 적용
boxes = selective_search.selective_search(image, mode='fast')

boxes_filter = selective_search.box_filter(boxes, min_size=20, topN=80)

fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(image)
for x1, y1, x2, y2 in boxes_filter:
  bbox = mpatches.Rectangle((x1, y1), (x2-x1), (y2-y1), fill=False, edgecolor='green', linewidth=1)
  ax.add_patch(bbox)

plt.axis('off')
plt.show()
```
