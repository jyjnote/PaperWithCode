```python
import pycaret.classification as pc
import pandas as pd


# 데이터를 Pandas DataFrame으로 변환
train_df = pd.DataFrame(train)
test_df = pd.DataFrame(test)
target_df = pd.DataFrame({'target': target})

# 훈련 데이터와 타겟을 결합
train_combined_df = pd.concat([train_df, target_df], axis=1)

# StratifiedKFold 설정
stratified_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)

# PyCaret 환경 설정
exp = pc.setup(
    data=train_combined_df,
    target='target',
    session_id=123,  # 랜덤 시드 설정 (재현성을 위해)
    train_size=0.8,  # 훈련 데이터와 검증 데이터의 비율 설정
    fold_strategy=stratified_kfold,  # StratifiedKFold 교차 검증 설정
    verbose=True  # 출력 메시지를 활성화
)

# 다양한 모델 생성 및 비교
models_to_compare = [
    'lr',          # Logistic Regression
    'knn',         # K Neighbors Classifier
    'nb',          # Naive Bayes
    'dt',          # Decision Tree Classifier
    'svm',         # SVM - Linear Kernel
    'rbfsvm',      # SVM - Radial Kernel
    'gpc',         # Gaussian Process Classifier
    'mlp',         # MLP Classifier
    'ridge',       # Ridge Classifier
    'rf',          # Random Forest Classifier
    'qda',         # Quadratic Discriminant Analysis
    'ada',         # Ada Boost Classifier
    'gbc',         # Gradient Boosting Classifier
    'lda',         # Linear Discriminant Analysis
    'et',          # Extra Trees Classifier
    'lightgbm'    # Light Gradient Boosting Machine
]

# 모델 비교
best_model = pc.compare_models(include=models_to_compare)

# 모델 학습
final_model = pc.finalize_model(best_model)

# 테스트 데이터에 대한 예측
predictions = pc.predict_model(final_model, data=test_df)

# 예측 결과 확인
predictions.to_csv('titanic_predictions(pycaret).csv', index=False)

# 모델 평가
print(pc.pull())
```
