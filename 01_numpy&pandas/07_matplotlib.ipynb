# box plot
- 4분위수 기준으로 데이터의 최소, 최대, 25%(Q1), 50%(Q2), 75%(Q3) 등 통계적인 부분들을 손쉽게 시각화
- 이상치를 빠르게 판단 가능
- IQR(Inter Quantile Range): Q3 - Q1 (3사분위수와 1사분위수의 차이)
- Box plot 최소값: Q1 - IQR * 1.5
- Box plot 최대값: Q3 + IQR * 1.5
- Box plot의 최소와 최대값을 벗어나면 이상치로 해석!

````python
q1, q2, q3 = df["팁금액"].quantile([0.25, 0.5, 0.75]).to_numpy()
q1, q2, q3
````

````python
iqr = q3 - q1
box_min = q1 - iqr*1.5
box_max = q3 + iqr*1.5
box_min, box_max
````

````python
y_ticks = np.array([q1, q2, q3,box_min, box_max])
y_ticks
````

````python
fig, ax = plt.subplots()
ax.boxplot(df["팁금액"]) # 옆으로 보기 => vert=False
ax.set_yticks(y_ticks)
plt.show()
````

````python
y_ticks=df[df["팁금액"]>box_max]['팁금액']
fig, ax = plt.subplots()
ax.boxplot(df["팁금액"]) # 옆으로 보기 => vert=False
ax.set_yticks(y_ticks)
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame df and '팁금액' column (replace with your actual DataFrame)
# df = pd.DataFrame({'팁금액': [your_data_here]})

# Assuming box_max is defined and represents the maximum value for the boxplot
box_max = df['팁금액'].max()

# Define the y-ticks
y_ticks = df[df['팁금액'] > box_max]['팁금액']

fig, ax = plt.subplots()
ax.boxplot(df['팁금액'], vert=False)  # Horizontal boxplot

# Set custom y-ticks
ax.set_yticks(range(len(y_ticks)))
ax.set_yticklabels(y_ticks)

# Rotate labels if needed
plt.yticks(rotation=45)

plt.show()
````
