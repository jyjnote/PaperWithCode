```
고유한 고객번호로만 이루어진 데이터프레임을 만들어서 features 변수에 넣으시오.
컬럼명은 ID로 하시오.
컬럼이 ID만 있는 데이터프레임을 만들라는 의미
중복 고객번호가 있으면 안된다.
```

````python
features = pd.DataFrame(df["ID"].unique(), columns=["ID"])
features
````

```
groupby 메소드를 이용하여 고객별로 구매가격의 평균을 구하고 데이터 프레임 형식으로 정돈 하고 구매가격의 컬럼명을 변경하여
features 변수의 데이터프레임과 left merge 하시오.
```

````python
agg_dict = {
    "구매가격":[("구매가격_평균","mean")]
}
df.groupby("ID").agg(agg_dict).reset_index()
````

````python
agg_dict = {
    "구매가격":[("구매가격_평균","mean")]
}
df.groupby("ID").agg(agg_dict).reset_index()

tmp = df.groupby("ID").agg(agg_dict)
tmp.columns = tmp.columns.droplevel()
tmp = tmp.reset_index()
tmp
````

````python
agg_dict = {
    "구매가격": "mean"
}
tmp = df.groupby("ID").agg(agg_dict).reset_index().rename(columns = {"구매가격": "구매가격_평균"})
features = pd.merge(features, tmp, how="left", on="ID")
features
````

```
groupby 메소드를 이용하여 고객별로 구매가격의 최대값과 대분류의 고유값의 개수(다양성)를
데이데이터 프레임 형식으로 정돈 하고 구매가격과 대분류의 컬럼명을 변경하여
features 변수의 데이터프레임과 left merge 하시오
```

````python
agg_dict = {
    "구매가격": "max",
    "대분류": "nunique"
}

tmp = df.groupby("ID").agg(agg_dict).reset_index().rename(columns = {"구매가격": "구매가격_최대값", "대분류": "대분류_다양성"})
features = pd.merge(features, tmp, how="left", on="ID")
features
````

```
groupby 메소드를 이용하여 고객별로 구매가격의 최소값과, 총 구매액
그리고 지점코드의 고유값 개수와 최빈값을 구하시오
구하는 과정에서 집계 컬럼에 대한 컬럼명도 정의하시오.
데이터프레임형식으로 정돈하고 features 변수의 데이터프레임과 left merge 하시오.
```

````python
agg_dict = {
    "구매가격" : [
        ("구매가격_최소값", lambda x: x[x>0].min() ), # 환불을 제외하고!
        ("총구매액", "sum")
    ],
    "지점코드" : [
        ("지점_다양성", "nunique"),
        ("주방문지점", lambda x : x.mode()[0] )
    ]
}
df.groupby("ID").agg(agg_dict)
````

````python
agg_dict = {
    "구매가격" : [
        ("구매가격_최소값", lambda x: x[x>0].min() ), # 환불을 제외하고!
        ("총구매액", "sum")
    ],
    "지점코드" : [
        ("지점_다양성", "nunique"),
        ("주방문지점", lambda x : x.mode()[0] )
    ]
}

tmp = df.groupby("ID").agg(agg_dict)
tmp.columns = tmp.columns.droplevel()
tmp = tmp.reset_index()
features = pd.merge(features, tmp, how="left", on="ID")
features
````
