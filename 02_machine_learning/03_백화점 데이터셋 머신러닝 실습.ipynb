````python
import numpy as np
import pandas as pd
````

````python
import os

DATA_PATH = os.getcwd() 
train_ft = pd.read_csv(f"{DATA_PATH}\\data\\store_train.csv")
train_ft
````

````python
tmp = pd.read_csv(f"{DATA_PATH}\\data\\features.csv")
tmp
````

````python
train_ft = train_ft.merge(tmp,how = "left", on="ID")
train_ft
````

````python
mask = train_ft.isnull().sum() > 0
train_ft.isnull().sum()[mask]
````

````python
train_ft = train_ft.fillna(0)
````

````python
train_ft.isnull().sum().sum()
````

````python
train_ft_cp=train_ft.select_dtypes(exclude=['object', 'category'])
train_ft_cp=train_ft_cp.iloc[:,1:]
train_ft_cp
````

````python
from sklearn.preprocessing import MinMaxScaler

min_scaler = MinMaxScaler()

train_ft_cp[train_ft_cp.columns]=min_scaler.fit_transform(train_ft_cp)
````

````python
target=train_ft['target']
target
````

````python
from sklearn.model_selection import train_test_split
# 학습용 입력데이터, 검증용 입력데이터, 학습용 정답데이터, 검증용 정답데이터
x_train, x_valid, y_train, y_valid  = train_test_split(train_ft_cp, target, test_size=0.2, random_state=42)
````

````python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=42)
model.fit(x_train, y_train)
````

````python
model.predict(x_valid),model.predict(train_ft_cp)
````

````python
from sklearn.metrics import accuracy_score
pred_1=model.predict(x_valid)
pred_2=model.predict(train_ft_cp)
accuracy_score(y_valid, pred_1),accuracy_score(target, pred_2)
````
