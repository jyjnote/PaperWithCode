```python
import numpy as np
import pandas as pd
import os

os.chdir('C:\\NLP_Papers_Review\\extra_part_01_Auto_ML\\data')
DATA_PATH=os.getcwd()


train=np.load(DATA_PATH+'/processed_train_data(titanic).npy')
test=np.load(DATA_PATH+'/processed_test_data(titanic).npy')
target=pd.read_csv(r'C:\NLP_Papers_Review\extra_part_01_Auto_ML\data\titanic_train.csv')['survived'].to_numpy()

train.shape,test.shape,target.shape
```

### 2. 환경 변수 설정하기
    Java_HOME 설정:

    Windows에서는 다음 단계를 따라 환경 변수를 설정합니다:
    제어판 > 시스템 및 보안 > 시스템 > 고급 시스템 설정을 클릭합니다.
    시스템 속성 창에서 환경 변수(N)... 버튼을 클릭합니다.
    시스템 변수 섹션에서 새로 만들기 버튼을 클릭하고, 변수 이름에 JAVA_HOME, 변수 값에 Java 설치 경로를 입력합니다 (예: C:\Program Files\Java\jdk-11.0.11).
    PATH 변수에 Java 추가:

    시스템 변수 섹션에서 Path 변수를 찾아 선택하고 편집 버튼을 클릭합니다.
    새로 만들기 버튼을 클릭하고 JAVA_HOME\bin을 입력합니다 (예: C:\Program Files\Java\jdk-11.0.11\bin).

```python
import h2o
from h2o.automl import H2OAutoML
import pandas as pd

# H2O 클러스터 시작
h2o.init()

# NumPy 배열을 pandas DataFrame으로 변환
train_df = pd.DataFrame(train)
test_df = pd.DataFrame(test)
target_df = pd.DataFrame({'target': target})

# 훈련 데이터와 타겟 결합
train_combined_df = pd.concat([train_df, target_df], axis=1)

# H2O 프레임으로 변환
h2o_train = h2o.H2OFrame(train_combined_df)
h2o_test = h2o.H2OFrame(test_df)

# 타겟 변수 및 특성 변수 지정
target_column = 'target'  # 타겟 열의 이름을 정의합니다
features = list(h2o_train.columns)
features.remove(target_column)

# H2O 프레임에 타겟 변수의 데이터 타입 설정 (이진 분류의 경우)
h2o_train[target_column] = h2o_train[target_column].asfactor()

# 모델 학습을 위한 H2OAutoML 설정
automl = H2OAutoML(max_models=20, seed=1234)
automl.train(x=features, y=target_column, training_frame=h2o_train)

# 모델을 사용하여 예측 수행
predictions = automl.leader.predict(h2o_test)

# 예측 결과를 Pandas DataFrame으로 변환
predictions_df = predictions.as_data_frame()

# 예측 결과를 CSV 파일로 저장
predictions_df.to_csv('C:\\NLP_Papers_Review\\extra_part_01_Auto_ML\\data\\titanic_predictions_h2o.csv', index=False)

print("예측 완료 및 'titanic_predictions_h2o.csv'로 저장되었습니다.")
```

```python
import h2o
from h2o.automl import H2OAutoML
import pandas as pd

# H2O 클러스터 시작
h2o.init()

# NumPy 배열을 H2O 프레임으로 변환
train_df = pd.DataFrame(train)
test_df = pd.DataFrame(test)
target_df = pd.DataFrame({'target': target})

# 훈련 데이터와 타겟 결합
train_combined_df = pd.concat([train_df, target_df], axis=1)

# H2O 프레임으로 변환
h2o_train = h2o.H2OFrame(train_combined_df)
h2o_test = h2o.H2OFrame(test_df)

# 타겟 변수 및 특성 변수 지정
target_column = 'target'
features = list(h2o_train.columns)
features.remove(target_column)

# H2O 프레임에 타겟 변수의 데이터 타입 설정 (이진 분류의 경우)
h2o_train[target_column] = h2o_train[target_column].asfactor()

# 모델 학습을 위한 H2OAutoML 설정
automl = H2OAutoML(max_models=10, seed=1234, max_runtime_secs=1800)  # 최대 30분 동안 학습
automl.train(x=features, y=target_column, training_frame=h2o_train)

# 모델을 사용하여 예측 수행
predictions = automl.leader.predict(h2o_test)

# 예측 결과를 Pandas DataFrame으로 변환
predictions_df = predictions.as_data_frame()

# 예측 결과를 CSV 파일로 저장
predictions_df.to_csv('C:\\NLP_Papers_Review\\extra_part_01_Auto_ML\\data\\titanic_predictions_h2o.csv', index=False)

print("예측 완료 및 'titanic_predictions_h2o.csv'로 저장되었습니다.")
```
