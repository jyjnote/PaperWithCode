# 컴퓨터 비전 모델 발전 연도표

## 1. AlexNet (2012)
- **의의**: 딥러닝이 컴퓨터 비전 분야에서 큰 혁신을 일으키게 한 모델로, ImageNet 대회에서 첫 번째로 딥러닝 기반 CNN을 사용하여 우승했습니다.
- **특징**: 8개의 레이어로 구성되어 있으며, ReLU 활성화 함수, Dropout, Data Augmentation 등을 도입하여 과적합을 방지하고 학습 속도를 높였습니다.
- **사용 예시**: 이미지 분류, 물체 인식, 이미지 검색 엔진 등에서 사용.

## 2. VGG16 (2014)
- **의의**: 단순한 구조를 통해 깊은 네트워크의 중요성을 강조하며, 더 깊은 신경망을 사용하는 방식이 성능을 높인다는 것을 보여주었습니다.
- **특징**: 16개의 레이어를 사용하며, 3x3의 작은 필터를 사용해 깊이를 증가시켰습니다. 모든 레이어가 동일한 크기의 필터를 사용하여 규칙성을 유지했습니다.
- **사용 예시**: 전이 학습을 통해 다양한 이미지 분류 및 물체 인식 작업에 활용.

## 3. RCNN (2014)
- **의의**: 물체 인식을 위한 초기의 딥러닝 모델로, 이미지에서 물체를 검출하는 데 큰 기여를 했습니다.
- **특징**: 선택적 검색(Selective Search)을 사용하여 이미지에서 제안된 물체 후보 영역을 추출하고, 각 후보 영역을 CNN을 통해 분류합니다.
- **사용 예시**: 객체 탐지 및 분류, 페이스북의 자동 이미지 태깅 시스템.

## 4. GoogleNet (2014)
- **의의**: 네트워크의 깊이와 폭을 동시에 확장할 수 있는 새로운 아키텍처인 인셉션(Inception)을 도입하여, 성능과 효율성을 동시에 달성했습니다.
- **특징**: 네트워크의 각 레이어에서 여러 크기의 필터를 병렬로 적용하는 인셉션 모듈을 도입하였으며, 파라미터 수를 크게 줄였습니다.
- **사용 예시**: 이미지 분류, 물체 탐지, 이미지 캡셔닝 등.

## 5. ResNet (2015)
- **의의**: 매우 깊은 신경망에서도 학습이 가능하도록 잔차 연결(residual connection)을 도입하여, 딥러닝의 한계를 극복했습니다.
- **특징**: 152 레이어의 깊은 네트워크를 사용하면서도, 잔차 연결로 인해 기울기 소실 문제를 해결하여 높은 성능을 달성했습니다.
- **사용 예시**: 이미지 인식, 물체 탐지, 분할, GAN 등의 다양한 비전 작업에 사용.

## 6. Faster RCNN (2015)
- **의의**: RCNN의 속도 문제를 개선하여, 실시간에 가까운 속도로 물체 탐지가 가능하게 했습니다.
- **특징**: 제안 영역을 생성하는 RPN(Region Proposal Network)과 물체를 분류하는 네트워크를 통합하여 효율성을 높였습니다.
- **사용 예시**: 자율주행 차량의 물체 인식, 비디오 분석, 실시간 물체 탐지.

## 7. UNet (2015)
- **의의**: 의료 영상 분할에서 큰 성과를 거둔 모델로, 특히 제한된 데이터셋에서도 좋은 성능을 발휘할 수 있습니다.
- **특징**: 대칭적인 U자 형태의 구조로, 인코더-디코더 방식으로 세밀한 분할 결과를 제공합니다.
- **사용 예시**: 의료 영상 분석, 생물학적 이미지 분석, 위성 이미지 분할.

## 8. SSD (2016)
- **의의**: 실시간 객체 탐지를 목표로 설계된 모델로, 빠른 속도와 높은 정확도를 동시에 제공합니다.
- **특징**: CNN의 출력층에서 여러 크기의 물체를 동시에 예측할 수 있는 피처 맵(feature map)을 사용하여 연산 속도를 높였습니다.
- **사용 예시**: 자율주행차, 감시 시스템, 모바일 어플리케이션에서의 실시간 객체 탐지.

## 9. GoogleNet V4 (2016)
- **의의**: 인셉션 모듈을 더욱 발전시켜 성능을 높였으며, ResNet의 잔차 개념을 결합해 네트워크의 효율성을 극대화했습니다.
- **특징**: 인셉션 모듈의 복잡성을 줄이면서도, 다양한 규모의 특징을 효과적으로 학습할 수 있도록 개선했습니다.
- **사용 예시**: 이미지 분류, 물체 탐지, 비디오 분석.

## 10. DeepLab V3 (2017)
- **의의**: 이미지 분할 작업에서 최고의 성능을 자랑하는 모델로, 다양한 크기의 객체를 효과적으로 분할할 수 있습니다.
- **특징**: 공액된 다중-분류기(atrous convolution)와 깊이 분리형 합성곱을 사용하여, 세밀한 공간 정보를 유지하면서도 효율적인 계산이 가능합니다.
- **사용 예시**: 자율주행차의 도로 상황 인식, 의료 이미지 분할, 인공지능 드론의 환경 인식.

## 11. Sniper (2018)
- **의의**: 큰 이미지에서도 작은 객체를 효과적으로 탐지할 수 있도록 고안된 모델로, 특히 스케일 변화를 효과적으로 다룹니다.
- **특징**: 복잡한 배경과 다양한 크기의 객체가 있는 환경에서도 높은 정확도를 유지하면서 빠른 속도로 객체를 탐지합니다.
- **사용 예시**: 도시 감시, 위성 이미지 분석, 스포츠 경기 분석.

## 12. YOLOv3 (2018)
- **의의**: 객체 탐지에서 실시간 성능을 목표로 한 모델로, 높은 정확도와 빠른 속도를 동시에 제공합니다.
- **특징**: 이미지의 여러 스케일에서 물체를 탐지할 수 있으며, 하나의 신경망을 통해 객체 위치와 클래스를 동시에 예측합니다.
- **사용 예시**: 자율주행차, 드론, 보안 시스템, CCTV 감시.

## 13. EfficientNet (2019)
- **의의**: 모델의 크기, 깊이, 너비를 동시에 최적화하여, 더 적은 자원으로 높은 성능을 달성하는 모델입니다.
- **특징**: 복잡한 아키텍처 검색 방법을 사용하여, 서로 다른 스케일의 네트워크를 동시에 최적화합니다.
- **사용 예시**: 모바일 기기에서의 이미지 인식, 대규모 이미지 분류, 객체 탐지.

## 14. ViT (Vision Transformer) (2020)
- **의의**: Transformer 아키텍처를 이미지 처리에 적용한 최초의 모델로, CNN을 사용하지 않고도 높은 성능을 달성했습니다.
- **특징**: 이미지 패치를 입력으로 받아들이고, 각각의 패치에 대한 특성을 Transformer를 통해 학습합니다. 대규모 데이터에서 특히 높은 성능을 보입니다.
- **사용 예시**: 이미지 분류, 객체 탐지, 비디오 분석, 자연어 처리와의 융합 작업.
