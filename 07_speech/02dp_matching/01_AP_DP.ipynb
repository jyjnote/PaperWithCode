### 얼라이언트 문제 (Alignment Problem)

얼라이언트 문제는 음성 인식에서 **입력된 음성 신호**와 **단어 또는 음소 등의 텍스트 표현**을 어떻게 맞추느냐에 대한 문제입니다. 음성 신호는 시간이 지남에 따라 연속적으로 변하는 특성을 가지고 있기 때문에, 입력된 음성과 출력 텍스트 사이에 정확한 매핑을 찾아야 합니다.

이 문제를 해결하려면 **음성 신호의 각 프레임을 어떤 음소나 단어에 맞추는지**에 대한 정렬 작업이 필요합니다. 예를 들어, 텍스트로 주어진 문장 "hello world"가 주어졌을 때, 이 문장을 음성 신호로 변환할 때 각 음소를 신호의 특정 부분에 맞추는 작업을 수행해야 합니다.

### 동적 계획법 (Dynamic Programming, DP)

동적 계획법(DP)은 주어진 문제를 작은 부분 문제로 나누어 해결하는 기법으로, 음성 인식에서의 얼라이언트 문제를 해결하는 데 널리 사용됩니다. 특히 **힌튼(HMM, Hidden Markov Model)** 기반 음성 인식에서 **가장 가능성 높은 단어 순서를 찾는** 데 사용됩니다.

DP는 **음성 신호의 각 시간 단계**와 **단어 또는 음소** 간의 **최적 경로**를 찾는 데 유용합니다. 이를 위해 **Viterbi 알고리즘**을 사용하여 가능한 모든 경로 중 가장 높은 확률을 가진 경로를 찾아냅니다.

#### Viterbi 알고리즘
Viterbi 알고리즘은 주어진 음성 신호에 대해 **최적의 상태 경로**를 찾는 DP 알고리즘입니다. 이 알고리즘은 각 상태에서의 **최적 확률**을 계산하고, 음성 신호가 특정 상태로부터 어떻게 발생했는지 추적하는 방식으로 작동합니다.

Viterbi 알고리즘을 음성 인식에서 적용할 때, 각 **음소**나 **단어**가 **특정 시간에 해당하는 음성 프레임**에 맞는지를 결정하는 데 사용됩니다. 이를 통해 텍스트와 음성 간의 정렬을 정확하게 수행할 수 있습니다.

### 얼라이언트 문제 해결을 위한 DP 활용 예시
1. **음소를 시간 축에 맞추는 문제**: 음성 신호에서 각 음소에 해당하는 프레임을 정확히 맞추는 과정은 DP를 통해 해결됩니다. 이를 통해 음성 신호가 어떤 음소로 변환되는지 정렬할 수 있습니다.
2. **단어를 맞추는 문제**: 발음 사전과 모델을 바탕으로, DP를 사용하여 입력된 음성에 대해 가장 가능성 높은 단어 조합을 찾아냅니다.

### 요약
- **얼라이언트 문제**는 음성 신호와 텍스트 간의 정확한 정렬을 찾는 문제로, **Viterbi 알고리즘**과 같은 DP 기법을 통해 해결할 수 있습니다.
- **동적 계획법(DP)**은 주어진 문제를 부분 문제로 나누어 해결하는 기법으로, 음성 인식 시스템에서 텍스트와 음성의 정렬을 최적화하는 데 사용됩니다.


### 음성 인식에서 유클리드 유사도와 MFCC

#### 1. MFCC (Mel-frequency cepstral coefficients)
MFCC는 음성 인식에서 가장 많이 사용되는 특징 벡터 중 하나입니다. 이는 **인간의 청각 시스템**의 주파수 응답을 모델링한 것이며, 음성 신호에서 **주요 특징**을 추출하는 데 사용됩니다.

- **MFCC 계산 과정**:
  1. **프레임 분할**: 음성 신호를 일정한 길이의 프레임으로 나눕니다.
  2. **윈도잉 (Windowing)**: 각 프레임에 윈도우 함수를 적용하여 주파수 분석을 할 수 있도록 합니다.
  3. **푸리에 변환**: 각 프레임에 대해 푸리에 변환을 통해 주파수 성분을 분석합니다.
  4. **Mel 필터 뱅크 적용**: Mel 주파수 스케일로 변환하여 음성의 중요한 주파수 성분을 강조합니다.
  5. **DCT (Discrete Cosine Transform)**: 주파수 성분을 더욱 압축하고 특징 벡터로 변환합니다.

MFCC는 **음성 신호의 패턴을 간단하고 압축된 형태**로 나타내므로 음성 인식에서 매우 유용합니다.

## MFCC 벡터 간의 유사도 계산

MFCC 벡터 $mfcc_p$와 $mfcc_q$는 각각 $d$ 차원의 특징 벡터입니다. 두 벡터 간의 유사도를 측정하는 여러 방법이 있습니다. 그 중에서 가장 기본적인 방법은 **유클리드 거리(Euclidean distance)** 또는 **코사인 유사도(Cosine similarity)**를 사용하는 것입니다.

### 1. 유클리드 거리(Euclidean Distance)

두 MFCC 벡터 간의 유사도를 유클리드 거리로 계산할 수 있습니다. 유클리드 거리는 두 벡터 간의 직선 거리를 의미하며, 다음과 같이 계산됩니다.

$$
d_{eucl}(mfcc_p, mfcc_q) = \sqrt{\sum_{i=1}^{d} (mfcc_p(i) - mfcc_q(i))^2}
$$

여기서 $mfcc_p(i)$는 벡터 $mfcc_p$의 $i$번째 요소이고, $mfcc_q(i)$는 벡터 $mfcc_q$의 $i$번째 요소입니다.

### 2. 코사인 유사도(Cosine Similarity)

코사인 유사도는 두 벡터 간의 각도를 기반으로 유사도를 측정합니다. 두 벡터의 코사인 각도가 0에 가까울수록 유사하다고 판단합니다. 코사인 유사도는 다음과 같이 계산됩니다.

$$
\text{cosine\_similarity}(mfcc_p, mfcc_q) = \frac{\sum_{i=1}^{d} mfcc_p(i) \cdot mfcc_q(i)}{\sqrt{\sum_{i=1}^{d} mfcc_p(i)^2} \cdot \sqrt{\sum_{i=1}^{d} mfcc_q(i)^2}}
$$

이 값은 $-1$과 $1$ 사이의 값을 가지며, 값이 $1$에 가까울수록 두 벡터는 매우 유사하다고 볼 수 있습니다.

### 3. 두 벡터 간 유사도의 해석

- **유클리드 거리**: 두 벡터가 가까울수록 유사도가 높고, 거리가 멀수록 유사도가 낮습니다. 이 값은 항상 0 이상입니다.
- **코사인 유사도**: 두 벡터가 같은 방향을 가질수록 유사도가 높고, 반대 방향일수록 유사도가 낮습니다. 값이 1에 가까우면 매우 유사하며, 0에 가까울수록 유사도가 낮습니다.


#### 3. MFCC와 유클리드 유사도 활용

음성 인식 시스템에서는 주어진 음성 신호의 **MFCC 특징 벡터**를 추출하고, 이를 **이미 저장된 음성 데이터의 MFCC 벡터**와 비교하여 가장 유사한 음성을 찾는 방식으로 인식이 이루어집니다. 유클리드 유사도는 이 비교에서 중요한 역할을 합니다.

##### 예시: MFCC 특징 추출 및 유클리드 유사도 계산

1. **입력 음성 신호에서 MFCC 특징 추출**:
   음성 신호를 처리하여 MFCC 특징 벡터를 추출합니다.

2. **저장된 음성 데이터와 비교**:
   저장된 음성 데이터에서 MFCC 벡터를 준비하고, 입력 음성의 MFCC 벡터와 **유클리드 거리**를 계산하여 가장 유사한 음성을 찾습니다.

3. **최소 거리 선택**:
   유클리드 거리 값이 가장 작은 음성을 선택하여 인식 결과를 도출합니다.

## DP 매칭을 이용한 실제 음성 문장 비교 예시

### 1. 음성 문장 예시

두 음성 문장을 비교해 봅시다. 음성 문장 A와 음성 문장 B는 다음과 같습니다.

- **음성 문장 A**: "I am learning"
- **음성 문장 B**: "I am studying"

각 음성 문장은 여러 개의 프레임으로 나눠지며, 각 프레임은 **MFCC (Mel-frequency cepstral coefficients)** 등의 특징 벡터로 표현될 수 있습니다. 

예를 들어, 문장 "I am learning"과 "I am studying"의 프레임들은 각각 다음과 같이 나눠집니다.

- **음성 문장 A (I am learning)**: 
  - $a_1$ = "I"
  - $a_2$ = "am"
  - $a_3$ = "learning"

- **음성 문장 B (I am studying)**: 
  - $b_1$ = "I"
  - $b_2$ = "am"
  - $b_3$ = "studying"

### 2. 유사도 행렬 생성

이제, DP 매칭을 위해 각 프레임에 대한 유사도를 계산합니다. 여기서는 **유클리드 거리**나 **코사인 유사도**를 사용하여 각 프레임 쌍 사이의 유사도를 구합니다.

|   | $b_1$ ("I") | $b_2$ ("am") | $b_3$ ("studying") |
|---|-------------|--------------|---------------------|
| **$a_1$ ("I")** | 0.1         | 0.3          | 0.7                 |
| **$a_2$ ("am")** | 0.2         | 0.0          | 0.6                 |
| **$a_3$ ("learning")** | 0.4         | 0.5          | 0.0                 |

위의 표는 각 음성 문장 A와 B의 프레임 간의 유사도를 나타냅니다. 예를 들어, $a_1$ ("I")와 $b_1$ ("I")는 매우 유사하므로 유사도가 0.1로 계산됩니다.

### 3. DP 행렬 생성

DP 매칭에서는 각 프레임 쌍에 대해 누적 비용을 계산하여 최적 경로를 찾습니다. 이때 비용은 유사도 행렬의 값을 사용하여 누적됩니다.

#### 초기 DP 행렬:

|   | $b_1$ ("I") | $b_2$ ("am") | $b_3$ ("studying") |
|---|-------------|--------------|---------------------|
| **$a_1$ ("I")** | 0.1         | 0.4          | 1.1                 |
| **$a_2$ ("am")** | 0.3         | 0.2          | 0.8                 |
| **$a_3$ ("learning")** | 0.7         | 0.7          | 0.3                 |

위의 DP 행렬에서는 첫 번째 프레임부터 마지막 프레임까지 각각의 비용을 누적한 값을 보여줍니다. 예를 들어, $a_1$ ("I")와 $b_1$ ("I")의 유사도가 0.1이므로 첫 번째 cell에는 0.1이 저장됩니다. 그 후, 각 프레임에 대해 최적 경로를 찾기 위해 누적 비용을 계산합니다.

### 4. 경로 추적

DP 매칭은 마지막 행렬에서 최적의 경로를 찾는 방법입니다. 경로 추적은 **역추적** 방식으로 진행됩니다. 마지막 cell부터 시작하여, 최소 누적 비용을 가진 경로를 따라가면 됩니다.

#### 경로 추적:

1. 마지막 cell (DP 행렬의 우측 하단)에서부터 시작하여 최소 비용을 가지는 경로를 추적합니다.
2. 경로 추적 시, 각 음성 문장의 프레임을 매칭합니다. 예를 들어, $a_1$ ("I")와 $b_1$ ("I")가 매칭되고, $a_2$ ("am")와 $b_2$ ("am")가 매칭되는 방식입니다.

### 5. 최종 매칭 결과

최종적으로, DP 매칭을 통해 얻은 결과는 다음과 같습니다:

- $a_1$ ("I") → $b_1$ ("I")
- $a_2$ ("am") → $b_2$ ("am")
- $a_3$ ("learning") → $b_3$ ("studying")

DP 매칭을 통해 두 음성 문장 간의 프레임 매칭을 최적화했으며, 각 음성 문장의 단어들 간의 관계를 정확히 찾아냈습니다.

### 6. DP 매칭의 의미

- **유연한 매칭**: 음성 문장 간의 프레임 순서나 시간 차이를 고려하여 유연하게 매칭을 할 수 있습니다. 예를 들어, 문장 "I am learning"과 "I am studying"의 경우, 단어의 순서가 다르지만 최적 경로를 통해 매칭을 찾을 수 있습니다.
- **음성 인식 정확도 향상**: DP 매칭은 음성 인식에서 **단어 또는 음소** 단위의 최적 매칭을 찾는 데 도움을 주며, 문장 간의 차이를 잘 반영할 수 있습니다.


```python
import numpy as np
import matplotlib.pyplot as plt
import random

# 음성 인식 예시 문장
sentence1 = "i told you hello"
sentence2 = "i havent you hola"

# 두 문장의 길이
len1 = len(sentence1)
len2 = len(sentence2)

# DP 행렬 생성
dp_matrix = np.zeros((len1+1, len2+1))

# DP 초기화 (편집 거리 알고리즘)
for i in range(len1 + 1):
    dp_matrix[i][0] = i
for j in range(len2 + 1):
    dp_matrix[0][j] = j

# DP 계산 (편집 거리)
for i in range(1, len1 + 1):
    for j in range(1, len2 + 1):
        cost = 0 if sentence1[i-1] == sentence2[j-1] else 1
        dp_matrix[i][j] = min(dp_matrix[i-1][j] + 1, dp_matrix[i][j-1] + 1, dp_matrix[i-1][j-1] + cost)

# 흔들림 효과를 위한 함수
def add_shake(matrix, intensity=0.2):
    """행렬에 시각적 흔들림을 추가하는 함수"""
    shaken_matrix = matrix.copy()
    noise = np.random.uniform(-intensity, intensity, shaken_matrix.shape)
    shaken_matrix += noise
    return shaken_matrix

# 최적 경로를 찾는 함수 (Backtracking)
def find_optimal_path(dp_matrix):
    """최소 편집 거리 경로를 추적하여 반환"""
    path = []
    i, j = len(dp_matrix) - 1, len(dp_matrix[0]) - 1
    
    while i > 0 and j > 0:
        path.append((i, j))
        if dp_matrix[i-1][j] < dp_matrix[i][j-1] and dp_matrix[i-1][j] < dp_matrix[i-1][j-1]:
            i -= 1
        elif dp_matrix[i][j-1] < dp_matrix[i-1][j] and dp_matrix[i][j-1] < dp_matrix[i-1][j-1]:
            j -= 1
        else:
            i -= 1
            j -= 1
    
    path.append((i, j))
    path.reverse()  # 경로를 시작점부터 끝까지 나열
    return path

# 시각화
fig, ax = plt.subplots(figsize=(8, 6))

# 최적 경로를 찾음
optimal_path = find_optimal_path(dp_matrix)

# 애니메이션으로 DP 행렬을 흔들리게 나타내기
for _ in range(100):  # 100번의 프레임
    ax.clear()  # 이전 프레임 지우기
    shaken_dp = add_shake(dp_matrix)  # 행렬에 흔들림 추가
    ax.imshow(shaken_dp, cmap='viridis', interpolation='nearest')
    ax.set_title('DP Matrix with Shaking Effect')
    ax.set_xlabel('Sentence 2')
    ax.set_ylabel('Sentence 1')
    
    # 최적 경로를 빨간선으로 표시
    path_x, path_y = zip(*optimal_path)
    ax.plot(path_y, path_x, color='red', linewidth=2)

    plt.pause(0.05)  # 애니메이션 속도 조정

plt.show()
```
